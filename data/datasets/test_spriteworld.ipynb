```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

def create_scatter_color_plot(x, y, c, title):
    norm = mcolors.Normalize(vmin=0, vmax=0.3)

    plt.figure(figsize=(7, 5))

    # Create scatter plot
    sc = plt.scatter(x, y, c=c, cmap='viridis', s=10, edgecolor='none', norm=norm)

    # Set the scale of the axes to logarithmic
    plt.xscale('log')
    plt.yscale('log')

    # Add color bar
    cbar = plt.colorbar(sc)
    cbar.set_label('Slot Identifiability Score')

    # Labeling the axes
    plt.xlabel('Reconstruction Error')
    plt.ylabel('Compositional Contrast')


    # Title (optional)
    plt.title(title)

    # Show plot
    plt.show()

x = without_velocity[:, 1]
y = without_velocity[:, 2]
c = without_velocity[:, 3]
create_scatter_color_plot(x, y, c, 'Compositional contrast vs reconstruction loss vs SIS without velocity latent')
```
